Based on the statistical analysis in Question.5, identify (if there are any) and remove the
outliers from the numerical columns.

```python
z_scores = stats.zscore(df[numerical_columns])
threshold = 2
outliers_indices = np.where(np.abs(z_scores) > threshold)
outliers_indices
```

```python
df_no_outliers = df[(np.abs(z_scores) <= threshold).all(axis=1)]
df.shape
df_no_outliers.shape
```
